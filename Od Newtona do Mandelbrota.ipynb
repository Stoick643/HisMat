Pri iskanju *kompleksnih* ničel z Newtonovo metodo je prvi rezultat dal Arthur Cayley leta 1879:  
**Izrek** Naj ima kompleksen kvadraten polinom $f(z) = a^2+bz+c$ dve ničli $\alpha$ in $\beta$ v kompleksni ravnini. Naj bo $L$ pravokotnica na sredini daljice med $\alpha$ in $\beta$. Potem velja, da uporaba Newtonove metode za $f(z)$ da bazena atrakcije za $\alpha$ in $\beta$ ravno polravnini, določeni z $L$.

Zadeve so bistveno bolj zanimive pri kubičnih polinomih. Tedaj govorimo o **Newtonovih fraktalih**, ki so poseben primer Juliajevih množic, zato o tem več kasneje.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import math as mth

# The function used to generate the fractal.
def f(x):
    return ((x)**(3)-1)

# The derivative of the above function.
def fd(x):
    return (3*x**(2))
    
def roots():
    return np.array([1,0,0,-1])
    
"""
step = 0.00390625
"""
step = 0.00390625
multval = int(1/step)

rootar = np.roots(roots())
root1 = rootar[0]
root2 = rootar[1]
root3 = rootar[2]

newx = []
newy = []
newz = []
newr = []
colors = [[0 for x in range(multval*2)] for x in range(multval*2)]

def newton(f, fd, x):
    f_temp = f(x)
    fd_temp = fd(x)
    xk = 0
    while 1:
        if ((fd_temp)==0):
            newr.append(0)
            return 0
        dx = f_temp / (fd_temp)
        if abs(dx) < 10e-3 * (1 + abs(x)):
            if(np.abs((root1-(x-dx)))<10e-3):
                rootval = 1
            if(np.abs((root2-(x-dx)))<10e-3):
                rootval = 2
            if(np.abs((root3-(x-dx)))<10e-3):
                rootval = 3
            newr.append(rootval)
            return rootval
        x = x - dx
        f_temp = f(x)
        fd_temp = fd(x)
        xk = xk + 1

def frange2(start, stop, step, r):
    l = start
    while l < stop:
        newx.append(r)
        newy.append(l)
        z = complex(r,l)
        newz.append(z)
        newton(f,fd,z)
        l += step

def frange(start, stop, step):
    r = start
    count = 0
    while r < stop:
        frange2(start, stop, step, r)
        count += 1
        r += step

frange(-1,1,step)

rangeval = multval*4*multval        

for i in range(0,rangeval):
        x_value = int((newx[i]+1)*multval)
        y_value = int((newy[i]+1)*multval)
        if(newr[i]==0):
            colors[y_value][x_value] = [0.0, 0.0, 0.0]
        if(newr[i]==1):
            colors[y_value][x_value] = [1.0, 0.0, 0.0]
        if(newr[i]==2):
            colors[y_value][x_value] = [0.0, 1.0, 0.0]
        if(newr[i]==3):
            colors[y_value][x_value] = [0.0, 0.0, 1.0]

delta = step
x_g = y_g = np.arange(-1.0, 1.0, delta)
X, Y = np.meshgrid(x_g, y_g)

im = plt.imshow(colors, interpolation='bilinear', cmap=cm.RdYlGn, origin='upper', extent=[-1,1,-1,1])

plt.show()
```

### Kantorjeva množica




Kantorjeva množica $C$ je definirana kot

\begin{align*}
  C_0&=[0,1],\\
  C_1&=[0,\tfrac13]\cup [\tfrac23,1],\\
  C_2&=[0,\tfrac19]\cup [\tfrac29,\tfrac13]\cup [\tfrac23,\tfrac79]\cup [\tfrac89,1],\\[-6pt]
  &\ \,\vdots\\[-12pt]
  C&=\bigcap_{n=0}^\infty C_n.
\end{align*}


Ker velja
$C_n=\{0.d_1d_2d_3\dots\colon {d_j\in \{0,2\}}$ za $1\leq j\leq n$}, je množica realnih števil v intervalu $[0,1]$, ki imajo trojiški  zapis samo v števkah $0$ in $2$ na prvih $n$ mestih. Tako je $C$ množica vseh števil v intervalu $[0,1]$, ki imajo vsaj en trojiški zapis samo z $0$-lami and
$2$-kami. 

Kantorjeva množica je 
- neštevna (diagonalni argument)
- perfektna
- popolnoma nepovezana
- ima Lebesgueovo mero 0
- nikjer gosta

Velja zanimiv izrek:


Interval $[0,1]$ je zvezna slika Kantorjeve množice.
Definiramo $f: C \to [0,1]$ kot 
$$f(\sum_{k \in N}\frac{a_k}{3^k})=\sum_{k \in N}\frac{c_k}{2^k}$$


In posplošitev:
Vsak kompakten metričen prostor je zvezna slika Kantorjeve množice.





**Definicija:** Naj bo $\mathcal H(S)$ prostor, ki predstavlja kolekcijo kompaktnih podmnožic v $S \subset R^n$ in $S$ zaprta množica.  

**Definicija:** Iterativni funkcijski sistem (IFS) je definiran kot kot končni niz $(T_i)_{i=1}^n$, tako da velja $ T(A) = T_1(A) \cup T_2(A) \cup \dots \cup T_n(A) $.

Če velja, da je vsaka transformacija $T_i$ kontrakcija, je kontrakcija tudi transformacija $T$!  

*Primer 1*  
Na intervalu $[0,1]$ naj bo $T_1 = x/3$ in $T_2 = x/3 + 2/3$. Potem je $\{T_1, T_2\}$ IFS za Kantorjevo množico.

*Primer 2*  
Na enotskem kvadratu definiramo: 
$$ T_1(x,y) = (x/2, y/2), $$
$$ T_2(x,y) = (x/2 + 1/2, y/2), $$
$$ T_3(x,y) = (x/2 + 1/2, y/2 + 1/2) $$
in dobimo *trikotnik Sierpinskega*.





```python
import numpy as np
import pylab
from random import randint

def midpoint(point1, point2):
    return [(point1[0] + point2[0])/2, (point1[1] + point2[1])/2]

curr_point = [0,0]  # our seed value for the chaos game
                    # It can fall anywhere inside the triangle

# our equilateral triangle vertices
v1 = [0,0]
v2 = [1,0]
v3 = [.5,np.sqrt(3)/2]

# Plot 5000 points
for _ in range(1000):
    # choose a triangle vertex at random
    # set the current point to be the midpoint
    # between the previous current point and
    # the randomly chosen vertex
    val = randint(0,2)
    if val == 0:
        curr_point = midpoint(curr_point, v1)
    if val == 1:
        curr_point = midpoint(curr_point, v2)
    if val == 2:
        curr_point = midpoint(curr_point, v3)
    # plot the new current point
    pylab.plot(curr_point[0],curr_point[1],'m.',markersize=2)

pylab.show()
```

### Benoît Mandelbrot


Po francoskem Grand Prix-u je prišlo do polstoletnega stagniranja, vse dokler niso računalniki postali tako zmogljivi, da so se lahko prikazovale slike Fatoujevih and juliajevih množic. Najbolj zaslužen za to je Benoit Mandelbrot. Zanimivo, da mu je idejo dal njegov stric, Szolem Mandelbrojt, profesor matematike v Parizu. Torej še vedno 'francoska povezava', pe toliko bolj, ker je bil njegov (stričev) doktorski mentor Jacques Hadamard.
Definicija Mandelbrotove množice je pravzprav kratka in enostavna:  
To je množica vrednosti $c$ v kompleksni ravnini, za katere je orbita $0$ pod iteracijo kvadratne funkcije $ z_{n+1} = z_n^2 + c$
omejena oz. bolj formalno, za $f_c(z) = z^2 + c$ je Mandelbrotova množica definirana kot $ M = \{c \in C: f_c^n(0) \not\to\infty\}  $

***
***
Kratka biografija: Benoit Mandelbrot (1924 - 2010) je poljski Žid, ki je z družino leta 1936 emigriral v Francijo in je študiral v Parizu in kasneje v Kaliforniji, imel je tudi dvojno državljanstvo, od ZDA in Francije. Leta 1958 je začel 35-letno kariero v IBM-u, in dostop do zmogljivih računalnikov mu je omogočil, da je bil eden prvih, ki je opazil fraktalne vzorce in kar je vodilo do odkritja Mandelbrotove množice.  
Prispeval je k raznim področjem: statistična fizika, meteorologija, hidrologija, geomorfologija, nevrologija, lingvistika, računalniška grafika itd.  
Leta 2012 je izšla njegova avtobiografija [The Fractalist: Memoir of a Scientific Maverick](https://www.amazon.com/Fractalist-Memoir-Scientific-Maverick/dp/030738991X/ref=tmm_pap_title_0?_encoding=UTF8&qid=1490558517&sr=1-1).
***
***





```python
import numpy as np
import matplotlib.pyplot as plt
def mandelbrot( h,w, maxit=120 ):
    """Returns an image of the Mandelbrot fractal of size (h,w)."""
    y,x = np.ogrid[ -1.4:1.4:h*1j, -2:0.8:w*1j ]
    c = x+y*1j
    z = c
    divtime = maxit + np.zeros(z.shape, dtype=int)

    for i in range(maxit):
        z = z**2 + c
        diverge = z*np.conj(z) > 2**2            # who is diverging
        div_now = diverge & (divtime==maxit)  # who is diverging now
        divtime[div_now] = i                  # note when
        z[diverge] = 2                        # avoid diverging too much

    return divtime

plt.figure(figsize=(100, 100))
plt.imshow(mandelbrot(400, 400))
plt.show()

```

[Riemanova ploskev](https://sl.wikipedia.org/wiki/Riemannova_ploskev) je enorazsežna kompleksna mnogoterost. Riemannove ploskve si lahko predstavljamo kot »deformirano obliko« kompleksne ravnine. Lokalno blizu vsake točke izgledajo kot delčki kompleksne ravnine, toda globalna topologija je lahko popolnoma drugačna. Lahko izgleda kot sfera ali torus ali kot dvojica zlepljenih plošč.
Imenuje se po nemškem matematiku Bernhardu Riemannu (1826 – 1866).

Obstaja netrivialen izrek, da je preprosto povezana Riemannova ploskev conformno izomorfna točno eni od 3 ploskev:
- kompleksna ravnina $\mathbb{C}$
- odprt enotski disk, torej množica točk $z$, za katere velja $|z|^2 = x^2 + y^2 < 1$
- Riemannova sfera


*Definicija*    **Normalna družina**

Normalna družina je prekompaktna družina zveznih funkcij. Neformalno to pomeni, da funkcije v družini niso preveč narazen, ampak se nekako držijo skupaj. 

Bolj formalno, družina $\mathcal F$ zveznih funkcij $f$, definirana na kompletnem metričnem prostoru $X$ z vrednostmi v drugem kompletnem metričnem prostoru $Y$, se imenuje *normalna*, če ima vsako zaporedje iz $\mathcal F$ podzaporedje, ki uniformno konvergira na komkpaktnih podmnožicah v $X$ k zvezni funkciji iz $X$ v $Y$.

To pomeni, da za vsako zaporedje funkcij v $\mathcal F$ obstaja podzaporedje $f_n(x)$ in zvezna funkcija $f(x)$ iz $X$ v $Y$, tako da za vsako kompaktno podmnožico $K$ v $X$ velja:

$$ \lim_{n\to\infty}\sup_{x\in K} d_{Y}(f_n(x),f(x))=0 $$

kjer je $d_Y$ metrika v prostoru $Y$.

  
**Opomba** 
Velja, da je limita v konvergentnem zaporedju holomorfnih funkcij tudi *holomorfna* funkcija.

**Montelov izrek ** - 1. oblika  
Vsaka lokalno omejena družina holomorfnih funkcij je normalna.

**Montelov izrek ** - 2. oblika  
Naj bo $\mathcal F$ družina holomorfnih funkcij iz Riemannove ploskve $S$ v Riemannovo sfero, ki 'izpusti' (vsaj) tri različne vrednosti. Potem je družina $\mathcal F$ normalna, to je, zaprtje $\bar{\mathcal F} \subset Hol(S, R^1)$ je kompaktna množica.

*Definicija* **Fatoujeva in Juliajeva množica**

Naj bo $S$ kompaktna Riemannova ploskev, $f : S \to S$ holomorfna funkcija in naj bo $f^n$ njena n-ta iteracija. Domena normalnosti za družino iteracij $\{f^n\}$ se imenuje *Fatoujeva množica* in njen komplement *Juliajeva množica*.  
Neformalno, Fatoujeva množica sestoji iz točk z lastnostjo, da se vse bližnje točke obnašajo podobno pod ponavljajočo se iteracijo funkcije in Juliajeva množica sestoji iz točk, kjer poljubno mala perturbacija lahko povzroči drastične spremembe v zaporedju iteriranih vrednosti.

*Alternativna definicija*  
Naj bo $f(z)$ kompleksna racionalna funkcija, to je $f(z) = p(z)/q(z)$. Potem obstaja končno število odprtih množic $F_1, \dots, F_r$, ki so levo invariantni za $f(z)$ in dodatno velja:
- unija $F_i$-jev je gosta v ravnini
- $f(z)$ se obnaša regularno in na enak način v vsaki množici $F_i$  

Vsaka od množic $F_i$ se imenuje *Fatoujeva domena* in njihova unija je Fatoujeva množica  $F(f)$ za $f(z)$.

** Izrek **  
Vsaka Fatoujeva domena ima isto mejo, ki je posledično Juliajeva množica.


*Definicija* **Newtonov fraktal** je Juliajeva množica za meromorfno funkcijo $z \to z - \frac{p(z)}{p'(z)}$, ki je dana z Newtonovo metodo, za polinom $p(z)$.
